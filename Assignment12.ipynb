```python
class Money:
  def __init__(self, value):
    self.value = value

  def __str__(self):
    return f"${self.value:.2f}"

class Bills(Money):
  def __init__(self, value):
    super().__init__(value)

class One(Bills):
  def __init__(self):
    super().__init__(1.00)

class Five(Bills):
  def __init__(self):
    super().__init__(5.00)

class Ten(Bills):
  def __init__(self):
    super().__init__(10.00)

class Twenty(Bills):
  def __init__(self):
    super().__init__(20.00)

class Coins(Money):
  def __init__(self, value):
    super().__init__(value)

class Penny(Coins):
  def __init__(self):
    super().__init__(0.01)

class Nickel(Coins):
  def __init__(self):
    super().__init__(0.05)

class Dime(Coins):
  def __init__(self):
    super().__init__(0.10)

class Quarter(Coins):
  def __init__(self):
    super().__init__(0.25)


class Wallet:
  def __init__(self):
    self.money_list = []

  def add_money(self, money):
    self.money_list.append(money)

  def get_total_value(self):
    total = sum(money.value for money in self.money_list)
    return total

  def __str__(self):
    money_str = ', '.join(str(money) for money in self.money_list)
    return f"Wallet: {money_str}"

wallet1 = Wallet()
wallet1.add_money(One())
wallet1.add_money(Five())
wallet1.add_money(Penny())

wallet2 = Wallet()
wallet2.add_money(Ten())
wallet2.add_money(Quarter())

wallet3 = Wallet()
wallet3.add_money(Five())
wallet3.add_money(Dime())
wallet3.add_money(Dime())
wallet3.add_money(Twenty())

wallets = [wallet1, wallet2, wallet3]

total_money = 0
for wallet in wallets:
    total_money += wallet.get_total_value()

for i, wallet in enumerate(wallets, start=1):
    wallet_value = wallet.get_total_value()
    print(f"Wallet {i} - Total: ${wallet_value:.2f}")
    print(wallet)

print("\nTotal Amount in All Wallets:", f"${total_money:.2f}")

```

```python
class ElectronicDevice:
    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.price = price
        self.powered_on = False

    def power_on(self):
        self.powered_on = True
        print(f"{self.brand} {self.model} is powered on.")

    def power_off(self):
        self.powered_on = False
        print(f"{self.brand} {self.model} is powered off.")

class Smartphone(ElectronicDevice):
    def __init__(self, brand, model, price, operating_system):
        super().__init__(brand, model, price)
        self.operating_system = operating_system

    def make_call(self, number):
        if self.powered_on:
            print(f"{self.brand} {self.model} is calling {number}.")
        else:
            print(f"Turn on {self.brand} {self.model} before making a call.")

class Tablet(ElectronicDevice):
    def __init__(self, brand, model, price, screen_size):
        super().__init__(brand, model, price)
        self.screen_size = screen_size

    def run_app(self, app_name):
        if self.powered_on:
            print(f"{self.brand} {self.model} is running {app_name}.")
        else:
            print(f"Turn on {self.brand} {self.model} before running an app.")

class Laptop(ElectronicDevice):
    def __init__(self, brand, model, price, processor):
        super().__init__(brand, model, price)
        self.processor = processor

    def open_browser(self, url):
        if self.powered_on:
            print(f"{self.brand} {self.model} is browsing {url}.")
        else:
            print(f"Turn on {self.brand} {self.model} before opening the browser.")

iphone = Smartphone("Apple", "iPhone 12", 999.99, "iOS")
ipad = Tablet("Apple", "iPad Pro", 799.99, "12.9-inch")
macbook = Laptop("Apple", "MacBook Pro", 1499.99, "Intel Core i5")

iphone.power_on()
iphone.make_call("555-1234")
iphone.power_off()

ipad.power_on()
ipad.run_app("Note-taking app")
ipad.power_off()

macbook.power_on()
macbook.open_browser("example.com")
macbook.power_off()
```
